A diferencia de las series temporales civiles que operan con el calendario natural de 365 días, las series de activos financieros que cotizan en mercados centralizados (como los futuros del oro en el COMEX, ticker `GC=F`) deben ajustarse estrictamente a los **días hábiles de negociación** (*Trading Days*).

El estándar internacional de Wall Street fija este periodo anual en **252 días**, deducidos mediante el siguiente descarte matemático:

* **Días del calendario civil:** 365 días.
* **Cierre por fines de semana:** -104 días (52 semanas x 2 días de cierre).
* **Cierre por festivos oficiales (CME Group):** -9 días promedio (Año Nuevo, Acción de Gracias, Navidad, etc.).
* **Días reales de operación bursátil:** **252 días**.

```python
# # Longitud del período (por ejemplo, meses en un año)
window_long = 252

# Calculamos la media móvil centrada sobre el precio real del oro
data["trend_estimated"] = data["Precio_Oro"].rolling(window=window_long, center=True).mean()

# 3. Diseño y visualización de la serie temporal + tendencia
plt.figure(figsize=(12, 6))

# Dibujamos la serie original del oro
plt.plot(data.index, data['Precio_Oro'], label="Precio Oro Real (GC=F)", color='#1f77b4', linewidth=0.6, alpha=0.7)

# Dibujamos la tendencia estimada
plt.plot(data.index, data['trend_estimated'], label=f"Tendencia Suavizada (Ventana: {window_long}d)", color='#2ca02c', linewidth=2)

# Personalización formal del gráfico
plt.title(f"Descomposición del Oro: Serie Original vs. Tendencia Estimada (Periodo: {window_long})", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Precio de Cierre (USD)", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

Al analizar el comportamiento conjunto de las cuatro ventanas de media móvil centrada (10, 50, 75 y 252 días), podemos evaluar de forma precisa cómo se comportan los componentes de nuestra serie temporal según el tamaño del filtro aplicado:

* **Ventanas de 10 y 50 días:** Ofrecen los suavizados más sutiles. La ventana de 10 días se mantiene prácticamente mimetizada con el precio bruto, reteniendo casi todo el ruido diario. Por su parte, la de 50 días logra eliminar ese ruido de ultra-alta frecuencia, pero conserva con gran nitidez las oscilaciones intermedias y las ondas estacionales de la serie.
* **Ventana de 75 días:** En este punto, cabría pensar que el suavizado iba a ser radicalmente mayor que el de la ventana de 50 y algo menor que el de 252. Sin embargo, aunque la línea verde de 75 días absorbe y aplana de forma notable los valles y las crestas del medio plazo, todavía permite que se filtren ligeras ondulaciones cíclicas del mercado, actuando como un puente de transición.
* **Ventana de 252 días:** Con esta configuración nos encontramos con el mayor de los suavizados vistos hasta ahora. Es más, la serie está tan suavizada que ya no se aprecian las oscilaciones estacionales habituales del activo. Es prácticamente una curva continua y limpia; es decir, **tendencia macroeconómica pura**. 

Existen sutiles irregularidades a lo largo de esta última línea (como si la hubiéramos pintado a mano alzada) y esto se debe a la componente de ruido. Al promediar en una ventana tan amplia de 252 días, el residuo aleatorio del mercado a veces hace que el resultado sea un poquito mayor y otras un poquito menor. Como este componente es intrínsecamente aleatorio e impredecible, no obtenemos una curva geométrica perfecta, sino una **recta tendencial ruidosa**.

La razón por la que se ha perdido por completo la componente estacional en la ventana de 252 días es porque estamos promediando en un espacio temporal equivalente al periodo natural de la estacionalidad del mercado financiero: **un año bursátil completo**. Es decir, los ciclos estacionales y de oferta/demanda tienden a repetirse y compensarse cada 252 puntos de negociación, y esta es la razón por la que la oscilación desaparece del gráfico. En general, la estacionalidad se extirpa cuando promediamos exactamente en un periodo completo. Es cierto que este proceso deja un nivel equivalente al promedio de la serie en ese año, pero si la amplitud de la oscilación no varía bruscamente con el tiempo, el resultado tiende a estabilizarse. Por tanto, obtenemos una línea limpia que indica la tendencia pura de la serie original más una pequeña componente de nivel de la oscilación.

Pasamos ahora a extraer de la serie original esta tendencia que acabamos de calcular con la ventana deslizante de tamaño igual al periodo. Al hacerlo nos queda como resultado todo lo que no sea tendencia en la serie. Es decir, las componentes de estacionalidad y ruido.

## 3. Extraer componente estacional + ruido

```python
# Extraer componente estacional + ruido
data['seasonality_noise'] = data['Precio_Oro'] - data['trend_estimated']
```

```python

# Inicializar y configurar el lienzo de visualización
plt.figure(figsize=(12, 6))

# Graficamos usando el índice temporal (fechas) y la serie desestacionalizada
plt.plot(data.index, data['seasonality_noise'], label="Estacionalidad + Ruido", color='#9467bd', linewidth=0.6)

# Añadimos una línea de referencia en Y=0 (el punto de equilibrio de la tendencia)
plt.axhline(y=0, color='black', linestyle='--', alpha=0.6, linewidth=1)

# Personalización formal del gráfico
plt.title("Serie Temporal del Oro sin Tendencia (Detrended con Periodo: 252)", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Desviación en USD (Valor real - Tendencia)", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

En este caso, la gráfica muestra lo que queda tras eliminar la tendencia y se puede ver claramente que ahora nuestra serie ya no aumenta su nivel con una trayectoria creciente, sino que nos encontramos con una serie que va oscilando siempre en torno al valor 0. Es cierto que el tamaño de la oscilación va en aumento —apreciándose una expansión drástica de la amplitud en el tramo final entre 2025 y 2026— y la propia oscilación cíclica continúa aún presente y mezclada con el residuo.

El siguiente paso es extraer formalmente la componente estacional de forma aislada. Para ello, vamos a promediar para cada posición dentro del periodo (del día 1 al 252), todos los valores de todos los puntos que están situados en el mismo punto del ciclo anual a lo largo de los diferentes años del histórico. Este procedimiento nos va a permitir anular el ruido aleatorio por promediado y obtener la componente estacional pura.

```python
import numpy as np

window_long = 252 # Longitud del período

# Aislar estacionalidad promediando por periodo (ciclo estacional)
data["period"] = np.arange(len(data)) % window_long
seasonality_avg = data.groupby("period")["seasonality_noise"].mean()
data["seasonality_estimated"] = data["period"].map(seasonality_avg)
```

```python
# Mostrar grafica de tendencia
plt.figure(figsize=(12, 5))

# Serie Temporal más tendencia estimada
plt.plot(data.index, data['seasonality_estimated'], label="Componente Estacional (Patrón Anual)", color='#1f77b4', linewidth=1)

# Añadimos una línea de equilibrio en cero
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5, linewidth=1)

plt.title("Componente Estacional Aislado de la Serie del Oro (Periodo: 252 días)", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Efecto Estacional en USD", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

Se ha estimado la oscilación de un periodo completo (252 días bursátiles) y se ha replicado secuencialmente tantas veces como periodos anuales tiene la serie histórica. Eso es exactamente lo que se muestra en la gráfica anterior. Se puede apreciar que la forma que tiene la estacionalidad de esta serie es muy similar a la estructura de una función seno o coseno, describiendo un comportamiento ondulatorio armónico y regular. Cabe destacar que la forma de la onda podría ser completamente distinta y repetirse de igual manera a lo largo de toda la serie; en este caso, al consolidarse a partir de los promedios históricos anuales del mercado del oro, este es el resultado geométrico de la componente estacional. 

Si se hubiera empleado otro tamaño para la ventana deslizante de la tendencia que no fuera el periodo exacto (o un múltiplo directo de este), al extraer esta componente se capturaría una forma totalmente diferente a la que se observa en la gráfica. Esta distorsión ocurriría en mayor o menor medida dependiendo del comportamiento del activo y de la desviación matemática respecto del periodo original. Al haber estructurado todo el proceso de forma coherente con el periodo real de la estacionalidad anual (252 días), la forma que extraemos limpia de la serie es exactamente la de la oscilación estacional pura.

**Extracción del Residuo Impredecible:**
Pasamos ahora a extraer la componente de ruido final o el residuo tras haber aislado el resto de los componentes estructurados (tendencia y estacionalidad). Para ello, a nuestra serie previa desprovista de tendencia (`seasonality_noise`), le restamos esta componente estacional que acabamos de calcular y replicar en todos sus periodos. El resultado final será el ruido puro y duro del mercado, libre de patrones y ciclos.

## 4. Extracción Ruido como residuo

```python
# Extraer ruido como residuo
data["noise_estimated_add"] = data["seasonality_noise"] - data["seasonality_estimated"]
```

```python
# Mostrar grafica de tendencia
plt.figure(figsize=(12, 5))

# Serie Temporal más tendencia estimada
plt.plot(data.index, data['noise_estimated_add'], label="Residuo / Ruido Aleatorio", color='#e377c2', linewidth=0.5, alpha=0.8)

# Añadimos una línea de equilibrio en cero
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5, linewidth=1)

plt.title("Componente de Ruido (Residuo Final) de la Serie Original del Oro", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Valor del Residuo en USD", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

Se puede ver que efectivamente la serie ya no tiene prácticamente oscilación ni tendencia y todo lo que le queda es el ruido. Aun así, en este caso aún podemos ver algún resquicio de oscilación al principio de la serie y también vemos que, aunque no tenga tendencia, sí que parece tener una amplitud creciente.

Esta expansión de la varianza en el tramo final (fines de 2025 e inicios de 2026) confirma que el residuo ha capturado con éxito la heterocedasticidad del mercado, es decir, cómo la volatilidad cambia y se amplifica drásticamente ante situaciones de estrés financiero real, dejando el terreno completamente preparado para la fase de modelado predictivo.

```python

from statsmodels.tsa.seasonal import seasonal_decompose

# Realizamos la descomposición automática de la serie (Modelo Aditivo)
# Ajustamos el período a 252 días (año bursátil) en lugar del genérico 50
df_decomposed = seasonal_decompose(data['Precio_Oro'], model='additive', period=252)

# Graficamos la descomposición institucional
fig = df_decomposed.plot()
fig.set_size_inches(12, 9)

# Estilizamos los paneles
axes = fig.get_axes()
axes[0].set_title("Descomposición Estructural Automática del Oro (Statsmodels - Aditivo)", fontsize=13, fontweight='bold', pad=15)
for ax in axes:
    ax.tick_params(labelsize=9)
    ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()

# 3. Acceder a la tendencia limpia (eliminando los NaNs de los extremos causados por el filtro centrado)
tendencia = df_decomposed.trend.dropna()
```

El hecho de que la componente residual aún presente resquicios de estacionalidad y amplitudes claramente crecientes a lo largo de la línea temporal está directamente relacionado con la naturaleza del filtro aplicado. Hemos utilizado un modelo aditivo, el cual asume de forma rígida que la magnitud de las fluctuaciones estacionales y del ruido es constante e independiente del nivel absoluto del precio. Sin embargo, como el oro ha experimentado un crecimiento exponencial masivo (escalando desde los 1.300 USD en 2019 hasta romper el techo de los 5.000 USD en 2026), la variabilidad de la serie se ha amplificado de forma proporcional a su valor.

Esta particularidad es el indicio teórico clásico de que la serie temporal responde mejor a un **modelo multiplicativo**. Por tanto, el siguiente paso metodológico será ejecutar de nuevo el proceso de descomposición bajo este enfoque, con el objetivo de evaluar si logramos estabilizar la varianza de los residuos y consolidar una serie completamente estacionaria y libre de patrones remanentes.

# 4. Implementación del Modelo Multiplicativo y Aislamiento de Ciclos
## 4.1. Extracción de Estacionalidad y Ruido mediante División

```python
# Extraer componente estacional + ruido mediante división proporcional
# Sumamos 1 para estabilizar el cálculo ante posibles anomalías numéricas
data['seasonality_noise'] = (data['Precio_Oro']+1) / data['trend_estimated']
```

```python
# Inicializar y configurar el lienzo de visualización
plt.figure(figsize=(12, 6))

# Graficamos usando el índice temporal (fechas) y la serie desestacionalizada
plt.plot(data.index, data['seasonality_noise'], label="Estacionalidad + Ruido (Multiplicativo)", color='#1f77b4', linewidth=0.5)

# Personalización formal del gráfico
plt.title("Serie Temporal del Oro sin Tendencia (Detrended con Periodo: 252)", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Desviación en USD (Valor real - Tendencia)", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

El gráfico muestra la serie desprovista de su tendencia macroeconómica principal a través de una división proporcional. Como se puede observar, el algoritmo ha logrado estabilizar la media histórica en torno al valor base de 1.0 (desplazada por la constante de seguridad aplicada) y ha controlado la expansión de la varianza. Un hallazgo crítico es que, bajo este enfoque, la amplitud de la crisis de 2020 y el shock parabólico de 2025-2026 se equiparan visualmente, revelando que ambos eventos tuvieron un impacto porcentual relativo similar sobre el activo.

No obstante, la serie aún conserva ligeras derivas y tendencias locales de medio plazo (especialmente visibles en el tramo final de 2025), lo que confirma que el residuo no es ruido puro y retiene memoria cíclica. Por ello, el siguiente paso indispensable es calcular la componente estacional aislada para remover estas oscilaciones. De nuevo, estimaremos este patrón promediando las posiciones homólogas dentro de cada periodo de la serie (252 días bursátiles).

```python
window_long = 252 # Longitud del período

# Aislar estacionalidad promediando por periodo (ciclo estacional)
data["period"] = np.arange(len(data)) % window_long
seasonality_avg = data.groupby("period")["seasonality_noise"].mean()
data["seasonality_estimated"] = data["period"].map(seasonality_avg)
```

```python
# Mostrar grafica de tendencia
plt.figure(figsize=(12, 5))

# Serie Temporal más tendencia estimada
plt.plot(data.index, data['seasonality_estimated'], label="Componente Estacional (Patrón Anual)", color='#1f77b4', linewidth=1)


plt.title("Componente Estacional Aislado (Modelo Multiplicativo - Periodo: 252)", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Fecha", fontsize=10)
plt.ylabel("Factor Estacional Ponderado", fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```
